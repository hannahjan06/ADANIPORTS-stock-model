```python
import pandas as pd
RANDOM_STATE = 42

stock = pd.read_csv("datasets/ADANIPORTS.csv")
stock.head()
```

```python
stock['Date'] = pd.to_datetime(stock['Date'])
stock['Year'] = stock['Date'].dt.year
stock['Month'] = stock['Date'].dt.month
stock['Day'] = stock['Date'].dt.day
stock['DayOfWeek'] = stock['Date'].dt.dayofweek
stock.drop(['Date', 'Series', 'Trades'], axis=1, inplace=True)

stock.head()
```

```python
stock = stock[stock['DayOfWeek'] < 5]
stock['DayOfWeek'].value_counts()
```

```python
import matplotlib.pyplot as plt

stock.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
for i in stock.select_dtypes(include=['object']):
    print(f"{i}: {stock[i].unique()}")
```

```python
stock.loc[stock['Symbol'] == 'MUNDRAPORT', 'Symbol'] = 1
stock.loc[stock['Symbol'] == 'ADANIPORTS', 'Symbol'] = 0

stock['Symbol_grps'] = stock['Symbol'].map({1: 'MUNDRAPORT', 0: 'ADANIPORTS'})
```

```python
plt.bar(stock['Symbol_grps'].value_counts().index, stock['Symbol_grps'].value_counts().values)
plt.xlabel('Symbol')
plt.ylabel('Count')
plt.show()
```

```python
stock['DayOfWeek_bins'] = stock['DayOfWeek'].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday'})
stock['DayBin'] = pd.cut(stock['Day'], 
                         bins=[0, 10, 20, 31], 
                         labels=['Early', 'Mid', 'Late'])

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    else:
        return 'Post-Monsoon'

stock['Season'] = stock['Month'].map(get_season)
```

```python
bin_features = ['DayOfWeek_bins', 'YearBin', 'DayBin', 'Season']

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

stock['DayOfWeek_bins'].value_counts().plot(kind='bar', ax=ax[0])
ax[0].set_title('Day of Week')

stock['DayBin'].value_counts().plot(kind='bar', ax=ax[1])
ax[1].set_title('Day')

stock['Season'].value_counts().plot(kind='bar', ax=ax[2])
ax[2].set_title('Season')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

stock_features = stock.drop(['Close'], axis=1)
strat_cols = stock_features[['Symbol_grps', 'DayOfWeek_bins', 'DayBin', 'Season']].dropna()
stock_target = stock['Close']

stock_features = stock_features.loc[strat_cols.index]
stock_target = stock_target.loc[strat_cols.index]

X_train, x_test, y_train, y_test = train_test_split(stock_features, stock_target, test_size=0.2,
                                                    random_state=RANDOM_STATE, 
                                                    stratify=strat_cols
)

print(f"Actual set proportions: {stock['Symbol_grps'].value_counts() / len(stock)}")
print(f"Train set proportions: {X_train['Symbol_grps'].value_counts() / len(X_train)}")
```

```python
X_train_dropped = X_train.drop(strat_cols, axis=1)
X_test_dropped = x_test.drop(strat_cols, axis=1)
stock_dropped = stock.drop(strat_cols, axis=1)

drop_cols = [
    'Symbol', 'DayBin', 'Season', 'Week'
]

X_train_final = X_train_dropped.drop(columns=drop_cols, errors='ignore')
X_test_final = X_test_dropped.drop(columns=drop_cols, errors='ignore')
stock_final = stock_dropped.drop(columns=drop_cols, errors='ignore')

corr_matrix = stock_final.corr()
corr_matrix["Close"].sort_values(ascending=False)
print(X_train_final.columns)
```

```python
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
import numpy as np

log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log1p, feature_names_out="one-to-one"),
    StandardScaler()
)

numerical_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)

default_num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)

preprocessor = make_column_transformer(
    (log_pipeline, ["Open", "High", "Low", "Last", "Prev Close", "VWAP", "Turnover", "Deliverable Volume", "Volume"]),
    (numerical_pipeline, ["%Deliverble"]),
    remainder= default_num_pipeline
)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

svr_reg = make_pipeline(preprocessor, SVR())

param_grid = [
    {
        "svr__kernel": ["rbf"],
        "svr__C": [2000],
        "svr__gamma": [0.01]
    }
]
grid_cv = GridSearchCV(svr_reg, param_grid=param_grid, scoring='neg_root_mean_squared_error')
grid_cv.fit(X_train_final, y_train)
```

```python
grid_cv.best_params_
```

```python
grid_cv.best_estimator_.score(X_train_final, y_train)
```

```python
grid_cv.best_estimator_.score(X_test_final, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = make_pipeline(preprocessor, RandomForestRegressor())

param_grid = [
    {
        "randomforestregressor__n_estimators": [100],
        "randomforestregressor__max_depth": [10],
        "randomforestregressor__min_samples_split": [2],
        "randomforestregressor__min_samples_leaf": [2]
    }
]

grid_cv_forest = GridSearchCV(rf_reg, param_grid=param_grid, scoring='neg_root_mean_squared_error')
grid_cv_forest.fit(X_train_final, y_train)
```

```python
grid_cv_forest.best_estimator_.score(X_train_final, y_train)
```

```python
grid_cv_forest.best_estimator_.score(X_test_final, y_test)
```

```python
# The RandomForestRegressor performed better than the Support Vector Regression (SVR) in this case.
# Therefore, we can use the RandomForestRegressor for predicting stock prices.
print(f"The best estimator: {grid_cv_forest.best_params_}")
print(f"The best score: {-grid_cv_forest.best_score_}")
```
